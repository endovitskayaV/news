```python
from datetime import date
import holidays
from settings import DATA_PATH,RAW_PATH
import pandas as pd
from pandas import Series
import datetime
```

```python
h = holidays.RU(years=[2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]).items()
```

```python
df = pd.DataFrame(h, columns=['date', 'name'])
df.to_csv(DATA_PATH/'holidays.csv', index=False, sep=';', encoding='utf-8')
```

```python
dates=df['date']
```

```python
def holiday_fun(row:Series,dates:Series)->Series:
    row['is_holiday'] = row['publish_date'].date() in dates.values
    return row        
```

```python
df = pd.read_csv(RAW_PATH / "train.csv", index_col= 0, parse_dates=['publish_date'])
df = df.apply(lambda row: holiday_fun(row, dates), axis=1)
```

```python
df['hour'] = pd.to_datetime(df['publish_date']).dt.strftime("%H").astype(int)
df['minute'] = pd.to_datetime(df['publish_date']).dt.strftime("%M").astype(int)
df
```
